```python
#泰国
新案sql = """
SELECT
assign.assigned_sys_user_id AS 'id',
assign.mission_group_name as '组别',
assign.director_name as '主管姓名',
assign.group_leader_name as '组长',
count(DISTINCT assign.debtor_id) '新案个数',
round(SUM(assign.assign_principal_amount / 100),0) as '分案本金',
sum( cr.催回本金 + cr.展期费 ) AS '回款本金',
round(SUM(IF(date(assign.mission_log_create_at) = date(unassign.mission_log_create_at),cr.催回本金 + cr.展期费,0)),0) as '新案D1回款本金'
FROM
ods_fox_mission_log assign
LEFT JOIN ods_fox_mission_log_unassign_reason mlur ON mlur.assign_mission_log_id = assign.mission_log_id
LEFT JOIN ods_fox_mission_log unassign ON unassign.mission_log_id = mlur.mission_log_id 
AND unassign.mission_log_operator = 'unassign' 
AND unassign.mission_log_create_at > assign.mission_log_create_at 
AND unassign.mission_log_create_at >= '{0}'
LEFT JOIN (SELECT
    cr.mission_log_id,
    MAX( repay_date ) 还款日期,
    SUM( delay_amount)/100 展期费,
    IF(SUM( repaid_principal_amount ) > max(ml.assign_principal_amount),max(ml.assign_principal_amount),SUM( repaid_principal_amount ))/100 催回本金
FROM
    ods_fox_collect_recovery cr LEFT JOIN ods_fox_mission_log ml on cr.mission_log_id = ml.mission_log_id
WHERE
    repay_date >= '{0}'
    AND date(repay_date) <= '{1}'
    AND batch_num IS NOT NULL
GROUP BY
    cr.mission_log_id) cr 
ON assign.mission_log_id = cr.mission_log_id
left join (
SELECT sue.user_id,sue.dimission_date 
FROM
ods_fox_sys_user_extend sue
JOIN ods_fox_sys_user su ON sue.user_id = su.id
WHERE
sue.dimission_date IS NOT NULL 
AND su.del_flag = 1
AND su.del_date >= '2000-05-10'
) sul on assign.assigned_sys_user_id = sul.user_id
left join (
SELECT user_id,work_day,attendance_status_flag
FROM
ods_fox_collect_attendance_dtl 
WHERE
attendance_status_flag > 3
) dtl on assign.assigned_sys_user_id = dtl.user_id AND dtl.work_day = DATE (assign.mission_log_create_at)
left join ods_fox_asset a on assign.mission_log_asset_id = a.asset_id
LEFT join ods_fox_asset_quality aq on a.asset_item_number = aq.asset_item_number
WHERE 1 = 1 
AND assign.mission_log_operator = 'assign'
AND assign.mission_log_create_at >= '{0}'
AND date(assign.mission_log_create_at) <= '{1}'
AND (
    mission_log_unassign_reason IN ( "每天自动撤案", "其他",  "每月自动撤案","2天未跟进","职位变更") 
    OR (mission_log_unassign_reason = "结清撤案" and not(unassign.mission_log_create_at is not null and cr.催回本金 is null))     OR mission_log_unassign_reason IS NULL 
    OR (mission_log_unassign_reason = "离职撤案" AND assign.mission_log_create_at <sul.dimission_date)
    OR (mission_log_unassign_reason = "展期" AND DATE ( assign.mission_log_create_at ) != DATE ( unassign.mission_log_create_at )) 
    OR (mission_log_unassign_reason in ("短假撤案","长假撤案","旷工","稽核下线") AND attendance_status_flag IS NULL)
)
AND ((assign.assign_asset_late_days = 1 AND assign.mission_group_name LIKE "A%" ) 
    OR ( assign.assign_asset_late_days in (1,8) AND assign.mission_group_name LIKE "B1%" ) 
    OR ( assign.assign_asset_late_days in (1,15) AND assign.mission_group_name LIKE "B2%" ) 
    OR ( assign.assign_asset_late_days = -2 AND assign.mission_group_name LIKE "P%" ) 
)
GROUP BY
assign.assigned_sys_user_id,
assign.mission_group_name,
assign.group_leader_name,
assign.director_name
""".format(first_day,yesterday)
df1 = pd.read_sql(新案sql,conn_ods)

新案累计sql = """
SELECT
assign.assigned_sys_user_id AS 'id',
assign.mission_group_name as '组别',
assign.director_name as '主管姓名',
assign.group_leader_name as '组长',
count(DISTINCT assign.debtor_id) '新案个数',
round(SUM(assign.assign_principal_amount / 100),0) as '分案本金',
sum(cr.催回本金 + cr.展期费 ) AS '回款本金'


FROM
ods_fox_mission_log assign
LEFT JOIN ods_fox_mission_log_unassign_reason mlur ON mlur.assign_mission_log_id = assign.mission_log_id
LEFT JOIN ods_fox_mission_log unassign ON unassign.mission_log_id = mlur.mission_log_id 
AND unassign.mission_log_operator = 'unassign' 
AND unassign.mission_log_create_at > assign.mission_log_create_at 
AND unassign.mission_log_create_at >= '{0}'
LEFT JOIN (SELECT
    cr.mission_log_id,
    MAX( repay_date ) 还款日期,
    SUM( delay_amount)/100 展期费,
    IF(SUM( repaid_principal_amount ) > max(ml.assign_principal_amount),max(ml.assign_principal_amount),SUM( repaid_principal_amount ))/100 催回本金
FROM
    ods_fox_collect_recovery cr LEFT JOIN ods_fox_mission_log ml on cr.mission_log_id = ml.mission_log_id
WHERE
    repay_date >= '{0}'
    AND date(repay_date) <= '{1}'
    AND batch_num IS NOT NULL
GROUP BY
    cr.mission_log_id) cr 
ON assign.mission_log_id = cr.mission_log_id
left join (
SELECT sue.user_id,sue.dimission_date 
FROM
ods_fox_sys_user_extend sue
JOIN ods_fox_sys_user su ON sue.user_id = su.id
WHERE
sue.dimission_date IS NOT NULL 
AND su.del_flag = 1
AND su.del_date >= '2000-05-10'
) sul on assign.assigned_sys_user_id = sul.user_id
left join (
SELECT user_id,work_day,attendance_status_flag
FROM
ods_fox_collect_attendance_dtl 
WHERE
attendance_status_flag > 3
) dtl on assign.assigned_sys_user_id = dtl.user_id AND dtl.work_day = DATE (assign.mission_log_create_at)
left join ods_fox_asset a on assign.mission_log_asset_id = a.asset_id
LEFT join ods_fox_asset_quality aq on a.asset_item_number = aq.asset_item_number
WHERE 1 = 1 
AND assign.mission_log_operator = 'assign'
AND assign.mission_log_create_at >= '{0}'
AND date(assign.mission_log_create_at) <= '{1}'
AND (
    mission_log_unassign_reason IN ( "每天自动撤案", "其他",  "每月自动撤案","2天未跟进","职位变更") 
    OR (mission_log_unassign_reason = "结清撤案" and not(unassign.mission_log_create_at is not null and cr.催回本金 is null))     OR mission_log_unassign_reason IS NULL 
    OR (mission_log_unassign_reason = "离职撤案" AND assign.mission_log_create_at <sul.dimission_date)
    OR (mission_log_unassign_reason = "展期" AND DATE ( assign.mission_log_create_at ) != DATE ( unassign.mission_log_create_at )) 
    OR (mission_log_unassign_reason in ("短假撤案","长假撤案","旷工","稽核下线") AND attendance_status_flag IS NULL)
)
AND ((assign.assign_asset_late_days = 1 AND assign.mission_group_name LIKE "A%" ) 
    OR ( assign.assign_asset_late_days in (1,8) AND assign.mission_group_name LIKE "B1%" ) 
    OR ( assign.assign_asset_late_days in (1,15) AND assign.mission_group_name LIKE "B2%" ) 
    OR ( assign.assign_asset_late_days = -2 AND assign.mission_group_name LIKE "P%" ) 
)
GROUP BY
assign.assigned_sys_user_id,
assign.mission_group_name,
assign.group_leader_name,
assign.director_name
""".format(月初,yesterday)
df1_0 = pd.read_sql(新案累计sql,conn_ods)
```
